```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
initial = 2869

numSims = 500

volatility = .12
stockReturn = .20

numSteps = 180
```

```python
x = np.ndarray(shape=(numSteps,numSims), dtype=float)
y = np.ndarray(shape=(numSteps,numSims), dtype=float)


for j in range(0,numSims):
    y[0][j]=initial
    for i in range(1,numSteps):
        x[i][j] = i
        y[i][j] = y[i-1][j] + (y[i-1][j] * ((stockReturn/numSteps)+((volatility/35) * np.random.normal())))
```

```python

plt.suptitle('Monte Carlo: ' + str(numSims) + ' Lines, ' + str(numSteps) + ' Steps, ' + str(volatility*100) +
             '% volatility, ' + str(stockReturn*100) + '% return')
plt.xlabel ('Time Step')
plt.ylabel ('Asset Price')
plt.xlim(-numSteps/100,numSteps+numSteps/100)
plt.plot(x,y)
plt.savefig('MonteCarlo.png', dpi=500)
plt.show()
plt.hist(y[numSteps-1], bins='auto')
plt.savefig('Histogram.png', dpi=500)
plt.show()
```
